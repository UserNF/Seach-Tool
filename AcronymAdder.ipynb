```python
import tkinter as tk
from tkinter import messagebox

# Function to save the dictionary to a file
def save_dict():
    with open("data.txt", "w") as file:
        for key, value in items.items():
            file.write(f"{key}:{','.join(value)}\n")
    messagebox.showinfo("Info", "Dictionary saved successfully!")

# Function to load the dictionary from a file
def load_dict():
    try:
        with open("data.txt", "r") as file:
            for line in file:
                key, values = line.strip().split(":")
                items[key] = values.split(",")
                listbox.insert(tk.END, f"{key}: {values}")
    except FileNotFoundError:
        pass

# Function to add a key-value pair to the dictionary
def add_item():
    key = entry_key.get()
    value = entry_value.get()
    if key and value:
        if key in items:
            messagebox.showwarning("Warning", "Key already exists!")
        else:
            items[key] = value.split(",")
            listbox.insert(tk.END, f"{key}: {value}")
            entry_key.delete(0, tk.END)
            entry_value.delete(0, tk.END)

# Initialize the main window
root = tk.Tk()
root.title("Key-Value Dictionary")

# Dictionary to store items
items = {"FCU": ["E", "K"]}

# Entry widget to add keys
entry_key = tk.Entry(root, width=25)
entry_key.pack(pady=5)
entry_key.insert(0, "Enter key")

# Entry widget to add values
entry_value = tk.Entry(root, width=25)
entry_value.pack(pady=5)
entry_value.insert(0, "Enter values (comma-separated)")

# Button to add items
add_button = tk.Button(root, text="Add Item", command=add_item)
add_button.pack(pady=5)

# Listbox to display items
listbox = tk.Listbox(root, width=50, height=10)
listbox.pack(pady=10)

# Button to save the dictionary
save_button = tk.Button(root, text="Save Dictionary", command=save_dict)
save_button.pack(pady=5)

# Load the dictionary when the application starts
load_dict()

# Run the application
root.mainloop()
```
